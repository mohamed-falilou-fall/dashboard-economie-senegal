# **Ecosystème entrepreneurial du Sénégal, à la lumière de son environnement économique de 1960 à 2024. Par Mohamed Falilou Fall**

## Chargement and Pre-Processing

```python
#pip install pandas plotly seaborn matplotlib streamlit openpyxl
```

```python
import pandas as pd

# fichier principal de la World Bank
df_raw = pd.read_csv('/content/API_SEN_DS2_en_csv_v2_11156.csv', skiprows=4)

# les années (en colonnes)
year_cols = [str(year) for year in range(1960, 2024)]
df = df_raw[df_raw['Country Name'] == 'Senegal'].copy()

# Transformation long
df_long = df.melt(id_vars=['Indicator Name', 'Indicator Code'],
                  value_vars=year_cols,
                  var_name='Year', value_name='Value')

# Cleaning
df_long['Year'] = df_long['Year'].astype(int)
df_long.dropna(subset=['Value'], inplace=True)
```

## DASHBOARD 1 : Dynamique Économique Générale

```python
import plotly.express as px

indicateurs_economie = [
    "GDP (constant 2015 US$)",
    "GDP per capita (constant 2015 US$)",
    "Inflation, consumer prices (annual %)",
    "GDP growth (annual %)"
]

df_economie = df_long[df_long['Indicator Name'].isin(indicateurs_economie)]

fig = px.line(df_economie, x="Year", y="Value", color="Indicator Name",
              title="Dynamique économique du Sénégal 1960 à 2024 par Mohamed Falilou Fall",
              color_discrete_sequence=px.colors.qualitative.Dark2)
fig.show()
```

## DASHBOARD 2 : Environnement des affaires

```python
indicateurs_affaires = [
    "New business density (new registrations per 1,000 people ages 15-64)",
    "Time required to start a business (days)",
    "Cost of business start-up procedures (% of GNI per capita)"
]

df_affaires = df_long[df_long['Indicator Name'].isin(indicateurs_affaires)]

fig = px.line(df_affaires, x="Year", y="Value", color="Indicator Name",
              title="Climat des affaires au Sénégal de 1960 à 2024 par Mohamed Falilou Fall",
              color_discrete_sequence=px.colors.qualitative.Dark2)
fig.show()
```

### Analyse : Environnement des affaires

## DASHBOARD 3 : Ouverture économique et IDE

```python
indicateurs_ide = [
    "Foreign direct investment, net inflows (% of GDP)",
    "Trade (% of GDP)",
    "Exports of goods and services (% of GDP)"
]

df_ide = df_long[df_long['Indicator Name'].isin(indicateurs_ide)]

fig = px.area(df_ide, x="Year", y="Value", color="Indicator Name",
              title="Ouverture économique du Sénégal de 1960 à 2024 par Mohamed Falilou Fall",
              color_discrete_sequence=px.colors.qualitative.Dark2)
fig.show()
```

## DASHBOARD 4 : Capital humain et entrepreneuriat

```python
indicateurs_education = [
    "School enrollment, tertiary (% gross)",
    "Literacy rate, adult total (% of people ages 15 and above)",
    "Unemployment, youth total (% of total labor force ages 15-24)"
]

df_edu = df_long[df_long['Indicator Name'].isin(indicateurs_education)]

fig = px.line(df_edu, x="Year", y="Value", color="Indicator Name",
              title="Capital humain et entrepreneuriat de 1960 à 2024 par Mohamed Falilou Fall",
              color_discrete_sequence=px.colors.qualitative.Dark2)
fig.show()
```

## DASHBOARD 5 : Numérisation et Accès au marché



```python
indicateurs_numerique = [
    "Individuals using the Internet (% of population)",
    "Secure Internet servers (per 1 million people)",
    "Mobile cellular subscriptions (per 100 people)"
]

df_tech = df_long[df_long['Indicator Name'].isin(indicateurs_numerique)]

fig = px.line(df_tech, x="Year", y="Value", color="Indicator Name",
              title="Numérisation et accès au marché de 1960 à 2024 par Mohamed Falilou Fall",
              color_discrete_sequence=px.colors.qualitative.Dark2)
fig.show()
```
